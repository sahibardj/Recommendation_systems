```python
import numpy as np
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#to ignore warings if any occur
warnings.filterwarnings("ignore") 
```

```python
#load the file.
#the file is not csv, it is tab seperated file, so a sep property is added.

#taking column names from Instruction given 
coloumn_names=["user_id","item_id","rating","time"]
df=pd.read_csv("ml-100k/u.data",sep="\t",names=coloumn_names,header=None)
moviet=pd.read_csv("ml-100k/u.item",sep="\|")
moviet=moviet.iloc[:-1,0:2]
moviet.columns=["item_id","movie_title"]
moviet.head()
df=pd.merge(df,moviet, on="item_id")
df.head()


```

```python
#to get everage rating of all the movie present in data set
df.groupby("movie_title").mean()["rating"].sort_values(ascending=False).head()
```

```python
ratings=pd.DataFrame(df.groupby("movie_title").mean()["rating"])
ratings["count"]=pd.DataFrame(df.groupby("movie_title").count()["rating"])
ratings.head()
```

```python
#create a pivote table for better vitusalization
movie_matrix=df.pivot_table(index="user_id",columns="movie_title",values="rating")
movie_matrix
```

```python
#establishing a relationship btween all movies and starwars
starwar_user_rating=movie_matrix["Star Wars (1977)"]
```

```python
#craeting dataframe
similar_to_sw=pd.DataFrame(movie_matrix.corrwith(starwar_user_rating),columns=["corelation"])
```

```python
similar_to_sw.dropna(inplace=True)
```

```python
similar_to_sw.sort_values("corelation", ascending=False)
```

```python
#fileration of suggestion more then 100
the_suggestions=similar_to_sw.join(ratings["count"])
```

```python
the_suggestions[the_suggestions["count"]>100].sort_values("corelation", ascending= False)
```

```python
# maiking a funtion for predictin movie:
def suggest_movie(movie_name):
    movie_user_rating=movie_matrix[movie_name]
    suggestions=pd.DataFrame(movie_matrix.corrwith(movie_user_rating),columns=["corelation"])
    suggestions.dropna(inplace=True)
    suggestions.sort_values("corelation", ascending=False)
    the_suggestions=suggestions.join(ratings["count"])
    prediction=the_suggestions[the_suggestions["count"]>100].sort_values("corelation", ascending= False)
    return prediction
```

```python
p= suggest_movie("As Good As It Gets (1997)")
p.head()
```
